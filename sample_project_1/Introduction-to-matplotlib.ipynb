## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter.
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5, 5), facecolor=(1, 1, 1)) # (width, height), sets background to white so you get the axis.
# Or just save as jpg seems to work aswell

# 3. Plot the data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
# 5. Save & Show (you save the whole figure)
fig.savefig("Images/sample-plot.png")

```

## Making figures with Numpy arrays
We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplot

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use the same data and create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a bar plot from a dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()); # (x, y)
ax.set(title="Nut Butter Store",
       ylabel="Price",
       xlabel="Type");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Horisontal bar chart",
       ylabel="Type",
       xlabel="Price")
plt.show()
```

```python
# Make some data for Histogram. randn = normal distrubution
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplots option 1 
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
#Plot to each different axis

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2 ( list indexing)
fig, ax = plt.subplots(nrows=2,
                   ncols=2,
                   figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use?(pyplot vs matplotlib OO method)

* When plotting something quickly, ok to use pyplot method.
* When plotting something more advanced, use OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head(), len(over_50), len(heart_disease)
```

```python
over_50.head()
```

```python
# Pyplot Method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
# ax.set_xlim([45, 80]); # set X limit from 45-80
```

```python
## OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--",
           color="red");
```

```python
over_50.head(2)
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10),
                               nrows=2,
                               ncols=1,
                               sharex=True)

# Add data to ax0
scatter =  ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"])
# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        xlabel="Age",
        ylabel="Cholesterol")
# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")
# Add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");




#ax1 thalach add the data
scatter =  ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])
# Customize ax1
ax1.set(title="Heart disease and max heart rate",
        xlabel="Age",
        ylabel="Max heart rate")
# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")
#add meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");


#  Add a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10),
                               nrows=2,
                               ncols=1,
                               sharex=True)

# Add data to ax0
scatter =  ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"],
                       cmap="winter")
# Customize ax0
ax0.set(title="Heart disease and cholesterol levels",
        xlabel="Age",
        ylabel="Cholesterol")
ax0.set_xlim(50, 80) # change the x axis limits


# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add meanline
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");


#---------------------------------- AX1

#ax1 thalach add the data
scatter =  ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")
# Customize ax1
ax1.set(title="Heart disease and max heart rate",
        xlabel="Age",
        ylabel="Max heart rate")
ax1.set_ylim(60, 200) # Change y lim. lina upp linjerna bättre på grid

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

#add meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");


#  Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
